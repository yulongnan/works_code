```python
import os
import re
import pandas as pd 
import numpy as np
```

```python
# 所有学生的学号
xuehao_1, xuehao_2,xuehao_3,xuehao_4= [], [], [], []

# filenames = os.listdir('B机制181_毕设电子材料')
# print(filenames)
# print (re.findall(r"\d+",filenames[0]))

BJ = ['B机制181_毕设电子材料','B机制182_毕设电子材料','B机制183_毕设电子材料','B机制184_毕设电子材料']

for j_num in range(4):
    filenames = os.listdir(BJ[j_num])

    for i_num in range(len(filenames)):
        index_XH = re.findall(r"\d+",filenames[i_num])
        if j_num == 0:
            xuehao_1.append(index_XH[0])
        if j_num == 1:
            xuehao_2.append(index_XH[0])
        if j_num == 2:
            xuehao_3.append(index_XH[0])
        if j_num == 3:
            xuehao_4.append(index_XH[0])
print('xuehao_1 == \n',xuehao_1)
print('xuehao_2 == \n',xuehao_2)
print('xuehao_3 == \n',xuehao_3)
print('xuehao_4 == \n',xuehao_4)


```
